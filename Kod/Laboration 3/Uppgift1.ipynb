## 5) Index och slices

- Index hämtar ett värde (t.ex. första ordet)
- Slice hämtar flera värden (t.ex. de två första orden)

OBS: Python börjar räkna på 0!

Negativa index räknar från slutet av listan. Index -1 är till exempel sista elementet, -2 nästsista osv.

```python
fordon = ["buss", "tåg", "spårvagn", "färja"]

print("Index 0:", fordon[0])
print("Slice 1:3:", fordon[1:3])
```
