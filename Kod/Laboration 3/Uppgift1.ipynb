## 5) Index och slices

### Exempel

- Slice hämtar flera värden (t.ex. de två första orden)

Negativa index räkna från slutet av listan. Index -1 är till exempel sista elementet, -2 nästsista osv.

```python
fordon = ["buss", "tåg", "spårvagn", "färja"]

print("Index 0:", fordon[0])
print("Slice 1:3:", fordon[1:3])
```
